#PYTHON

```r
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn as sk
import matplotlib.pyplot as plt
```

```r
df = pd.read_csv("/content/StudentsPerformance.csv")
df
```

```r
df.info()
```

```r
df.describe()
```

```r
df.isnull().any()
```

```r
df['math score'].plot(kind='hist', bins=20, title='math score')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```r
df.groupby('gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```r
df.groupby('race/ethnicity').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```r
df.groupby('parental level of education').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```r
df.plot(kind='scatter', x='reading score', y='writing score', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```r
df['writing score'].plot(kind='line', figsize=(8, 4), title='writing score')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```r
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['race/ethnicity'].value_counts()
    for x_label, grp in df.groupby('gender')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('gender')
_ = plt.ylabel('race/ethnicity')
```

#R

```r
library(dplyr)
library(ggplot2)
```

```r
df <- read.csv("/content/StudentsPerformance.csv")
```

```r
#scatter plot of math score vs. reading scores, colored by gender
options(repr.plot.width = 10, repr.plot.height = 6)
ggplot(df, aes(x = math.score, y = reading.score, color = gender)) +
  geom_point() +
  labs(title = "Scatter Plot of Math vs. Reading Scores", x = "Math Score", y = "Reading Score") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```r
ggplot(df, aes(x = race.ethnicity, fill = gender)) +
  geom_bar(position = "stack") +
  labs(title = "Stacked Bar Plot of Race/Ethnicity by Gender", x = "Race/Ethnicity", y = "Count") +
   theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```r
# Bar chart of counts for 'parental level of education'
education_counts <- table(df$parental.level.of.education)

barplot(education_counts,
        main = "Bar Chart of Parental Level of Education",
        xlab = "Education Level", ylab = "Count",
        col = "skyblue",
        horiz = FALSE,  # Set to TRUE for horizontal bars
        width = 0.7,    # Adjust the width of the bars
        space = 0.2,    # Adjust the space between bars
        cex.names = 0.8  # Adjust the size of axis labels
)
```
